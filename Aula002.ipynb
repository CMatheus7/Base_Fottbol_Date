```python
import pandas as pd
import numpy as np
```

### Coletando as planilhas das ultimas 12 temporadas 1º da Premier League ###

```python
df1 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1314/E0.csv') # Campeonato Inglês 2013/2014
df2 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1415/E0.csv') # Campeonato Inglês 2014/2015
df3 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1516/E0.csv') # Campeonato Inglês 2015/2016
df4 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1617/E0.csv') # Campeonato Inglês 2016/2017
df5 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1718/E0.csv') # Campeonato Inglês 2017/2018
df6 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1819/E0.csv') # Campeonato Inglês 2018/2019
df7 = pd.read_csv('https://www.football-data.co.uk/mmz4281/1920/E0.csv') # Campeonato Inglês 2019/2020
df8 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2021/E0.csv') # Campeonato Inglês 2020/2021
df9 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2122/E0.csv') # Campeonato Inglês 2021/2022
df10 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2223/E0.csv') # Campeonato Inglês 2022/2023
df11 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv') # Campeonato Inglês 2023/2024
df12 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2425/E0.csv') # Campeonato Inglês 2024/2025
```

### Concatenando as Planilhas ###

```python
df = (df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12)
df = pd.concat(df)
```

```python
df = df [['Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR','B365H', 'B365D', 'B365A']]
```
