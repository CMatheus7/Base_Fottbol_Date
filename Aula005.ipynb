### Importando Bibliotecas ###

```python
import pandas as pd
import numpy as np
```

### Coletando as planilhas das ligas secundarias  ###

```python
# Argentina
df1 = pd.read_csv("https://www.football-data.co.uk/new/ARG.csv")

# Austria
df2 = pd.read_csv("https://www.football-data.co.uk/new/AUT.csv")

# Brazil
df3 = pd.read_csv("https://www.football-data.co.uk/new/BRA.csv")
```

### Concatenando as Planilhas ###

```python
df = (df1, df2, df3)
df = pd.concat(df)
```

### Visualizando o arquivo gerado ###

```python
df
```

### Selecionando as colunas  que serão utilizadas ###

```python
df = df [['Country','Date', 'Home', 'Away', 'HG', 'AG', 'Res','PSCH','PSCD','PSCA']]
```

### Renomeando as Colunas ###

```python
df.columns = ['Liga','Data','Mandante', 'Visitante', 'Gols_Mandande', 'Gols_Visitante', 'Resultado', 'Odds_Madante', 'Odds_Empate','Odds_Visitante']
```
